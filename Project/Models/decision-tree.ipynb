```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
```

### Importing dataset

```python
train_df = pd.read_csv('../input/ml-project-final-dataset/processedTraining.csv')
test_df = pd.read_csv('../input/ml-project-final-dataset/processedTesting.csv')
```

```python
test_SKID = test_df['sk_id_curr']
test_features = test_df.drop(['sk_id_curr'], axis=1)
```

```python
main_X = train_df.drop(['target'], axis=1)
main_y = train_df['target']
X_train, X_test, y_train, y_test = train_test_split(main_X, main_y, test_size = 0.3, random_state = 42)
```

### Trying out different models

```python
tree_model = DecisionTreeClassifier(random_state=1, class_weight = "balanced")
tree_model.fit(X_train, y_train)
y_pred = tree_model.predict(X_test)
f1_score(y_test, y_pred, average='macro')
```

```python
forest_model_grid = DecisionTreeClassifier(max_depth=27, random_state=1)
forest_model_grid.fit(X_train, y_train)
y_pred = forest_model_grid.predict(X_test)
f1_score(y_test, y_pred, average='macro')
```

```python
forest_model_grid = DecisionTreeClassifier(max_depth=27, random_state=1, class_weight = "balanced")
forest_model_grid.fit(X_train, y_train)
y_pred = forest_model_grid.predict(X_test)
f1_score(y_test, y_pred, average='macro')
```

#### Using some random values for parameter

```python
max_depth_list = [5, 7, 14, 19, 23, 27, 35, 40]

for i in max_depth_list :
    forest_model_grid = DecisionTreeClassifier(max_depth= i, random_state=1, class_weight = "balanced")
    forest_model_grid.fit(X_train, y_train)
    y_pred = forest_model_grid.predict(X_test)
    print("F1 score for depth = " + str(i) + " Score = ", f1_score(y_test, y_pred, average='macro'))
```

```python
max_features_list = ['sqrt', 'log2', None]
splitter_list = ['best', 'random']
for i in max_features_list:
    for j in splitter_list:
        forest_model_grid = DecisionTreeClassifier(max_depth= 27, random_state=1, class_weight = "balanced", max_features = i, splitter = j)
        forest_model_grid.fit(X_train, y_train)
        y_pred = forest_model_grid.predict(X_test)
        print("F1 score for max_reature = " + str(i) + " Splitter = " + str(j) + " Score = ", f1_score(y_test, y_pred, average='macro'))
```
