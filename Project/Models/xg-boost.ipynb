```python
import numpy as np
import pandas as pd 
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

### Importing DataSet 

```python
train_df = pd.read_csv('../input/ml-project-final-dataset/processedTraining.csv')
test_df = pd.read_csv('../input/ml-project-final-dataset/processedTesting.csv')
```

```python
test_SKID = test_df['sk_id_curr']
test_features = test_df.drop(['sk_id_curr'], axis=1)
```

```python
main_X = train_df.drop(['target'], axis=1)
main_y = train_df['target']
X_train, X_test, y_train, y_test = train_test_split(main_X, main_y, test_size = 0.3, random_state = 42)
```

### Model Implementation

```python
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test) 
f1_score(y_test, y_pred, average='macro')
```

```python
learning_rate_list = [1.2, 1.5, 1.8]
gamma_list = [5, 7, 9]

for i in learning_rate_list :
    for j in gamma_list:
        xgb_model = XGBClassifier(learning_rate = i, gamma = j, n_estimators=50, max_depth=29)
        xgb_model.fit(X_train, y_train)
        y_pred = xgb_model.predict(X_test)
        print("F1 score for learning_rate = " + str(i) + " gamma = "+ str(j) + " Score = ", f1_score(y_test, y_pred, average='macro'))
```

```python
subsample_list = [0.63, 0.65, 0.68]
for i in subsample_list:
    xgb_model = XGBClassifier(learning_rate = 1.5, gamma = 7, n_estimators=50, max_depth=29, subsample = i)
    xgb_model.fit(X_train, y_train)
    y_pred = xgb_model.predict(X_test)
    print("F1 score for subsample = " + str(i) + " Score = ", f1_score(y_test, y_pred, average='macro'))
```

```python

xgb_model_2 = XGBClassifier(n_estimators=500,max_depth=29, learning_rate = 1, gamma = 7)
xgb_model_2.fit(X_train, y_train)
y_pred = xgb_model_2.predict(X_test)
f1_score(y_test, y_pred, average='macro')
```

### Choosing best and training on whole training data

```python
final_model_r = XGBClassifier(learning_rate = 1, gamma = 7, n_estimators=50, max_depth=29)
final_model_r.fit(main_X, main_y)
test_pred = final_model_r.predict(test_features)
```

```python
test_df["target"] = test_pred
to_submit = test_df[["sk_id_curr", "target"]]
to_submit.to_csv('XGBoost.csv', index=False)
```
