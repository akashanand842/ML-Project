```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
import math
```

### Importing train and test preprocessed data

```python
tst = pd.read_csv(os.path.join('../input/mlprojectcolumnsintersection', "processedTesting.csv"))
```

```python
trn = pd.read_csv(os.path.join('../input/mlprojectcolumnsintersection', "processedTraining.csv"))
```

```python
tst.columns = tst.columns.str.lower()
trn.columns = trn.columns.str.lower()
```

### Converting the columns into list

```python
col_test = tst.columns.values.tolist()
col_test
```

```python
col_train = trn.columns.values.tolist()
col_train
```

### Intersection of train and test list

```python
lst = [value for value in col_train if value in col_test]
len(lst)
```

```python
train_cols = lst
train_cols.append('target')

test_cols = lst 
test_cols.append('sk_id_curr')
len(train_cols)
```

```python
len(test_cols)
```

```python
finalProcessedTest = tst[tst.columns.intersection(test_cols)]
finalProcessedTest.head()
```

```python
finalProcessedTrain = trn[trn.columns.intersection(train_cols)]
finalProcessedTrain.head()
```

```python
finalProcessedTest.to_csv('processedTesting.csv', index=False)
finalProcessedTrain.to_csv('processedTraining.csv', index=False)
```
